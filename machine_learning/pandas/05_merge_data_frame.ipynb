```python
import pandas as pd
```

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)
left
```

```python
right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
right
```

```python
result = pd.merge(left, right, on="key")
result
```

```python
result.to_parquet("result.parquet")
```

```python
result_from_io = pd.read_parquet("result.parquet")
result_from_io
```

```python
result.to_csv("result.csv")
```

```python
result_from_io = pd.read_csv("result.csv")
result_from_io
```
